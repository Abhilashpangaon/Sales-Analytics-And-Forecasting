```python
import pandas as pd

# Load data
df = pd.read_csv('../data/sales_data.csv')

# Convert date column
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# Check missing values
print(df.isnull().sum())

# Save cleaned dataset
df.to_csv('../data/cleaned_sales_data.csv', index=False)

print("Data cleaning completed")
```
